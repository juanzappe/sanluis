```python
import sys
from pathlib import Path
import csv 
import pandas as pd
import re

# Agregamos la carpeta src al path para poder importar
path_inicio = Path.cwd().parent/"src"
sys.path.append(str(path_inicio))

from utilidades.constantes import  TOTAL_PATH, TOTAL_FINAL_PATH
```

```python
# Leer el archivo CSV
df = pd.read_csv(TOTAL_PATH, sep=";", decimal=",", dtype=str)

# Quedarte solo con las columnas que te interesan
df_limpio = df[["Fecha de Emisión", "Tipo de Comprobante", "Imp. Total"]].copy()

# Convertir la fecha a formato datetime (opcional)
df_limpio["Fecha de Emisión"] = pd.to_datetime(df_limpio["Fecha de Emisión"], errors="coerce")

# Convertir el importe a número (quitando comas, puntos o símbolos)
df_limpio["Imp. Total"] = (
    df_limpio["Imp. Total"]
    .replace({r"[^0-9,.-]": ""}, regex=True)  # eliminar símbolos como $
    .str.replace(",", ".", regex=False)       # reemplazar coma decimal por punto
    .astype(float)
)

# Guardar el resultado limpio
df_limpio.to_csv(TOTAL_FINAL_PATH, index=False, sep=";")
print(df_limpio.head())
```
