```python
import sys
from pathlib import Path
import csv 
import pandas as pd
import re

# Agregamos la carpeta src al path para poder importar
path_inicio = Path.cwd().parent.parent/"src"
sys.path.append(str(path_inicio))

from main import unificar_ventas_excel_a_csv, parse_line
from utilidades.constantes import DATA_PATH, PATRON_MOSTRADOR_PATH, MOSTRADOR_SALIDA_PATH, PATRON_SERVICIOS_PATH, SERVICIOS_SALIDA_PATH
```

UNIFICAR TODOS LOS MESES DEL MOSTRADOR EN UN SOLO DATASET

```python
unificar_ventas_excel_a_csv(
    DATA_PATH,
    PATRON_MOSTRADOR_PATH,
    MOSTRADOR_SALIDA_PATH
)
```

UNIFICAR TODOS LOS SERVICIOS EN UN SOLO DATASET


```python
data = []
for ventas_file in DATA_PATH.glob(PATRON_SERVICIOS_PATH):
    with ventas_file.open(encoding="latin-1") as f:
        for i, line in enumerate(f, 1):
            parsed = parse_line(line, file_name=ventas_file.name, line_num=i)
            if parsed is not None:
                data.append(parsed)

df = pd.DataFrame(data)

if len(df) == 0:
    print("No se parseó ningún registro. Verificá el formato de los archivos.")
else:
    df["fecha_emision"] = pd.to_datetime(df["fecha_emision"], format="%Y%m%d", errors="coerce")
    df["fecha_emision"] = df["fecha_emision"].dt.strftime('%Y-%m-%d')
    df.to_csv(SERVICIOS_SALIDA_PATH, index=False, encoding="utf-8-sig")
    print(f"¡Listo! {len(df)} comprobantes unificados en {SERVICIOS_SALIDA_PATH}.")
    print("Columnas exportadas:", df.columns.tolist())
```
